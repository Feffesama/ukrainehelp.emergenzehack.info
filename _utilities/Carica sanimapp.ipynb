```python
template_title="per migranti promosso da GRIS - Gruppi immigrazione e Salute della Società Italiana di Medicina"

template_descr="Presidio sanitario per l'assistenza sanitaria a persone straniere sprovviste di permesso di soggiorno e ai cittadini comunitari in condizione di fragilità sociale e amministrativa, in 4 lingue, realizzata dal Gris Lazio in collaborazione con Sanità di Frontiera Onlus e Team Dev. Per ulteriori info vai al link del progetto Sanimapp."

template_link="https://www.sanitadifrontiera.org/sanimapp/"
```

```python
sanimapp=pd.read_csv("Sanimapp.csv")
```

```python
for i,row in sanimapp.fillna("").iterrows():
    title=row["Servizio"].strip() + " " + template_title  + " #" + row["Città"].strip() + " #Lazio"
    data={}
    data["Descrizione"]=template_descr
    data["Link"]=template_link
    data["Indirizzo"]=row["Indirizzo"].strip()
    data["Orari"]=row["Orari"].strip()
    data["Telefono"]=row["Numero di Telefono"].strip()
    data["Posizione"]=row["y"].replace(",",".") + " " + row["x"].replace(",",".") + " 0 0"
    data["Fonte"]="GRIS - Gruppi immigrazione e Salute della Società Italiana di Medicina"
    print(title)
    body='<pre><yamldata>'+yaml.dump(data)+'</yamldata></pre>'
    print(body)
    #print (dict(row))
    try:
        r.create_issue(title=title, body=body,labels=["Stranieri", "accettato", "Sanimapp"])
    except Exception as m:
        print(m)
        pass

    
```
